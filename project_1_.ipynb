````python
platform_sales = data.groupby('platform')['total_sales'].sum().reset_index()

# –®–∞–≥ 2: –û—Ç—Å–æ—Ä—Ç–∏—Ä—É–π—Ç–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã –ø–æ —Å—É–º–º–∞—Ä–Ω—ã–º –ø—Ä–æ–¥–∞–∂–∞–º
sorted_platforms = platform_sales.sort_values('total_sales', ascending=False)

# –í—ã–±–µ—Ä–∏—Ç–µ —É–∫–∞–∑–∞–Ω–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ—Ç–µ–Ω—Ü–∏–∞–ª—å–Ω–æ –ø—Ä–∏–±—ã–ª—å–Ω—ã—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º
num_top_platforms = 3
potential_profitable_platforms = sorted_platforms.head(num_top_platforms)

# –°–æ–∑–¥–∞–π—Ç–µ —Å—Ç–æ–ª–±—á–∞—Ç—É—é –¥–∏–∞–≥—Ä–∞–º–º—É –¥–ª—è —Å—É–º–º–∞—Ä–Ω—ã—Ö –ø—Ä–æ–¥–∞–∂ –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º
colors = ['blue', 'orange', 'green']  # –°–ø–∏—Å–æ–∫ —Ü–≤–µ—Ç–æ–≤
plt.figure(figsize=(12, 8))
plt.bar(potential_profitable_platforms['platform'], potential_profitable_platforms['total_sales'], color=colors)
plt.title('–°—É–º–º–∞—Ä–Ω—ã–µ –ø—Ä–æ–¥–∞–∂–∏ –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º', fontsize=15)
plt.xlabel('–ü–ª–∞—Ç—Ñ–æ—Ä–º–∞', fontsize=12)
plt.ylabel('–ú–∏—Ä–æ–≤—ã–µ –ø—Ä–æ–¥–∞–∂–∏', fontsize=12)
plt.show()
````

````python
current_year = 2016
last_years_data = data[data['year_of_release'] >= current_year - 3]

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ —Å—É–º–º–∞—Ä–Ω—ã—Ö –ø—Ä–æ–¥–∞–∂ –¥–ª—è –∫–∞–∂–¥–æ–π –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã
platform_sales = last_years_data.groupby('platform')['total_sales'].sum().reset_index()

# –°–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –ø–ª–∞—Ç—Ñ–æ—Ä–º –ø–æ —Å—É–º–º–∞—Ä–Ω—ã–º –ø—Ä–æ–¥–∞–∂–∞–º
sorted_platforms = platform_sales.sort_values('total_sales', ascending=False)

# –ü–æ–ª—É—á–µ–Ω–∏–µ —Ç–æ–ø-–ª–∏–¥–∏—Ä—É—é—â–∏—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º
top_platforms = sorted_platforms.head(5)['platform'].tolist()

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞ "—è—â–∏–∫ —Å —É—Å–∞–º–∏" –¥–ª—è –∫–∞–∂–¥–æ–π –∏–∑ —Ç–æ–ø-–ª–∏–¥–µ—Ä–æ–≤
plt.figure(figsize=(12, 8))
boxplot_data = []
labels = []
for i, platform in enumerate(top_platforms):
    platform_data = last_years_data[last_years_data['platform'] == platform]['total_sales']
    boxplot_data.append(platform_data)
    labels.append(f'Platform {i+1}')
    # –î–æ–±–∞–≤–ª–µ–Ω–∏–µ –Ω–∞–∑–≤–∞–Ω–∏—è –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã –æ–∫–æ–ª–æ –∫–∞–∂–¥–æ–≥–æ —è—â–∏–∫–∞
    plt.text(i+1, platform_data.median(), platform, ha='center', va='bottom')

plt.boxplot(boxplot_data, labels=labels)
plt.title('–ì–ª–æ–±–∞–ª—å–Ω—ã–µ –ø—Ä–æ–¥–∞–∂–∏ –∏–≥—Ä –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º –∑–∞ –ø–æ—Å–ª–µ–¥–Ω–∏–µ 3 –≥–æ–¥–∞', fontsize=15)
plt.xlabel('–ü–ª–∞—Ç—Ñ–æ—Ä–º–∞', fontsize=12)
plt.ylabel('–ú–∏—Ä–æ–≤—ã–µ –ø—Ä–æ–¥–∞–∂–∏ (–º–∏–ª–ª–∏–æ–Ω—ã)', fontsize=12)
plt.ylim(0, 3)
plt.show()
````

–í –Ω–∞—á–∞–ª–µ 2000 –±—ã–ª–æ –±–æ–ª—å—à–æ–µ —Ä–∞–∑–Ω–æ–æ–±—Ä–∞–∑–∏–µ –∫–æ–Ω—Å–æ–ª–µ–π –Ω–æ —Å–æ –≤—Ä–µ–º–µ–Ω–µ–º –∏—Ö –æ—Å—Ç–∞–ª–æ—Å—å –≥–æ—Ä–∞–∑–¥–æ –º–µ–Ω—å—à–µ. –í –æ—Å–Ω–æ–≤–Ω–æ–º –∂–∏–∑–Ω–µ–Ω–Ω—ã–π —Ü–∏–∫–ª –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã –¥–ª–∏—Ç—Å—è 8-10 –ª–µ—Ç.

—Ç–∞–∫ –∂–µ —Å—Ç–æ–∏—Ç—Ç —Ä–∞–∑—Ä–∞–±–∞—Ç—ã–≤–∞—Ç—å –∏–≥—Ä—ã –∂–∞–Ω—Ä–∞ shooter, action sports.–û–Ω–∏ —â–∞—Å –ø–æ–ø—É–ª—è—Ä–Ω—ã 

````python
#–£–∏–º–Ω–≥–æ–∂–∞–µ–º –æ—Ü–µ–Ω–∫–∏ –¥–ª—è –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ–≥–æ —Å—Ä–∞–≤–Ω–µ–Ω–∏—è 
last_years_data['user_score'] = last_years_data['user_score']*10
for platform in top_platforms:
    # –°–æ–∑–¥–∞–Ω–∏–µ –ø–æ–¥–Ω–∞–±–æ—Ä–∞ –¥–∞–Ω–Ω—ã—Ö —Ç–æ–ª—å–∫–æ –¥–ª—è —Ç–µ–∫—É—â–µ–π –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã
    subset = last_years_data[last_years_data['platform'] == platform][['critic_score', 'user_score', 'total_sales']]
    
    # –£–¥–∞–ª–µ–Ω–∏–µ —Å—Ç—Ä–æ–∫ —Å–æ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ NaN
    subset = subset.dropna()
    
    # –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏
    correlation_critic = subset['critic_score'].corr(subset['total_sales'])
    correlation_user = subset['user_score'].corr(subset['total_sales'])
    
    print("–ü–ª–∞—Ç—Ñ–æ—Ä–º–∞:", platform)
    print("–ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è —Å –æ—Ü–µ–Ω–∫–æ–π –∫—Ä–∏—Ç–∏–∫–æ–≤:", correlation_critic)
    print("–ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è —Å –æ—Ü–µ–Ω–∫–æ–π –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π:", correlation_user)
    print("------------------------")
    
    # –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –¥–∏–∞–≥—Ä–∞–º–º—ã —Ä–∞—Å—Å–µ—è–Ω–∏—è
    plt.scatter(subset['critic_score'], subset['total_sales'], color='blue', label='–û—Ü–µ–Ω–∫–∞ –∫—Ä–∏—Ç–∏–∫–æ–≤')
    plt.scatter(subset['user_score'], subset['total_sales'], color='green', label='–û—Ü–µ–Ω–∫–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π')
    plt.xlabel('–û—Ü–µ–Ω–∫–∏')
    plt.ylabel('–û–±—â–∏–µ –ø—Ä–æ–¥–∞–∂–∏ (–º–∏–ª–ª–∏–æ–Ω—ã)')
    plt.legend()
    plt.show()
````
